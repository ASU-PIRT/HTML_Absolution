### <span style="color: Chartreuse;">*New!*</span> <span style="color: darkorange;">HTML Line Breaks and Indentation</span>

The first step is going to be to turn all of the export code into readable code. To do so:

* Identify all element tags that require a line break.
    * These will be the tags for block elements.
    * This also be block and configuration void elements, such as `<!DOCTYPE html>` and `<meta>`.
        * Insert `\n` newline code in python after the closing character or tag.

* Develop method to identify nesting level and indent each newline accordingly.
    * Initiate variable `nest_tier=int(0)` that determines the nesting tier as an integer.
        * The value of the nesting tier will be determine the number of instances of non-breaking space characters to be inserted after every newline created multiplied by the variable `tier_indent=int(2)`.

    * For all block elements (except `<p>`) opening tags:
        1. add +1 to `nest_tier`
            * `nest_tier += 1`
        2. add a newline with `\n`
        3. insert `' '` by a multiple of `nest_tier`.

    * For all block element ***closing tags***:
        1. subtract -1 to `nest_tier`
            * `nest_tier -= 1`
        2. add a newline BEFORE the closing tag.
        3. add a newline AFTER the tag.
        4. insert `&nbsp` by a multiple of `nest_tier` after the closing newline from step 3.

Apply solution.

```python
# Lucky Character Locator !

def character_locator(char='<', string=text, index_start: int = 0, index_end: int = -1):

    '''
    Character Locator:
    - Finds ALL instances of a specific character or substring in a string.
    - Returns an ordered list with the indices of said character.

    Parameters:
    char: the character in the string you're searching for

    string: the string you're searching in.

    index_start: the index of the string to begin search from.

    index_end: the index of the string where the search will end. By default set to -1 to 
    indicate the final index of the string.
    '''

    index_list = []
    index = index_start # taken from function parameter
    loop = True


    while loop:
        index = string.find(char, index)
        if index == -1:
            loop = False
        else:
            index_list.append(index)
        index += 1

    return index_list


#index_list = character_locator('font-weight:700')
#print(index_list)
#len(index_list)
```

```python
def html_organize(text=text):
    
    '''
        This function uses the ".find()" method to find opening and closing angled brackets `<` and `>`.

            * The function searches the document in order from beginning to end, creating new lines and indents based on HTML block rules.

        The current version of this function assumes the following lists:

            * block_opener
            * block_closer
            * self_closer
        
        which are respectively comprehensive lists of HTML 

            * opening tag 'heads', eg. `<style`
            * closing tags, eg. '</style'
            * self-closing tag heads, eg. '<meta'

    '''
    
    #Initialize
    
    new_text = text #makes copy of text
    open_tag_index = new_text.find('<') #finds the first opening bracket

    
    #Initialize Conditional Parameters
    #index_end = -1 # will end loop when open_tag_index becomes this value #
    loop = True #switch for `while` loop 
    

    #Initialize Formatting Parameters

    spaces_per_indent: int = 2 #determines integer number of spaces for indentation in general. 4 is another popular option
    nest_tier: int = 0 # This keeps track of what tier of nesting you're in as an integer.
    indent = ' ' * int(spaces_per_indent) #When multiplied by the current "nest_tier" it yields the proper amount of indentation.

    newlines = 0 # When set to 1, a new block tag will create a new line for itself. 


    ## Begin Loop

    while loop:

        # Loop Breaker
        if open_tag_index == -1:
            loop = False
            break

        
        close_index = new_text.find('>', open_tag_index) #index of tag's closing bracket

        tag = new_text[open_tag_index : close_index+1] #complete tag, eg. <span>
        #print(tag) # diagnostics

    
        # The follow conditional creates a tag head from the current tag so that it matches the 'block_opener' list.
        if tag.find(' ') != -1:
            index_increment = tag.find(' ')
            tag_head = new_text[open_tag_index : (open_tag_index+index_increment)]

        else:
            tag_head = new_text[open_tag_index : close_index]


        
        # Formatting Option 1: Block Openers
        if tag_head in block_opener:

            #print(f'tag is block opener\n')

            #newline before opening tag
            new_text = new_text[0:open_tag_index] + ('\n'*newlines + indent*nest_tier*newlines) + new_text[open_tag_index:]
            
            #update open & close bracket indexes
            open_tag_index += len('\n'+indent*nest_tier)
            close_index = new_text.find('>', open_tag_index)

            #increase indent:
            nest_tier += 1

            #newline after opening tag
            new_text = new_text[0:close_index+1] + ('\n' + indent*nest_tier) + new_text[close_index+1:]

            #set newlines flag:
            newlines = 1


        # Formatting Option 2: Block Closers
        elif tag in block_closer:

            #print(f'tag is block closer\n')

            #decrease indent:
            nest_tier -= 1

            #newline before closing tag
            new_text = new_text[0:open_tag_index] + ('\n' + indent*nest_tier) + new_text[open_tag_index:]

            #update indexes
            #open_tag_index += len('\n'+indent*nest_tier)
            open_tag_index = new_text.find('<', open_tag_index)

            #print(f'close bracket index updated from {close_index} to {close_index + len('\n'+indent*nest_tier)}')
            #close_index += len('\n'+indent*nest_tier)
            close_index = new_text.find('>', open_tag_index)

            #newline after closing tag
            new_text = new_text[0:close_index+1] + ('\n' + indent*nest_tier) + new_text[close_index+1:]

            #set newlines flag:
            newlines = 1

        # Formatting Option 3: Self-closing block tags
        elif tag_head in self_closer:

            #print(f'tag is self closer')

            #newline before closing tag
            new_text = new_text[0:open_tag_index] + ('\n'*newlines + indent*nest_tier) + new_text[open_tag_index:]

            #update indexes
            #open_tag_index += len('\n'+indent*nest_tier)
            open_tag_index = new_text.find('<', open_tag_index)
            

            #print(f'close bracket index updated from {close_index} to {close_index + len('\n'+indent*nest_tier)}')
            #close_index += len('\n'+indent*nest_tier)            
            close_index = new_text.find('>', open_tag_index)
            

            #newline after closing tag
            new_text = new_text[0:close_index+1] + ('\n' + indent*nest_tier) + new_text[close_index+1:]

            #set newlines flag:
            newlines = 0 # Only option with 0, for now.

        open_tag_index = new_text.find('<', open_tag_index+1)

            
    return new_text

new_text = html_organize()
#print(new_text)
```

```python
## Formatting the CSS

def css_organize(text=new_text):

    #Copy Text
    new_text = text
    
    #-- style tags indices --

    #Find the style open tag:
    open_start = new_text.find('<style')
    open_end = new_text.find('>', open_start)
    open_tag = new_text[open_start : open_end+1]

    #Find the style close tag index:
    close_start = new_text.find('</style>')

    #Getting the CSS text itself
    css = new_text[open_end + 1 : close_start]

    #--------------

    #Finding the Indent for style tag
    open_indent_start = new_text.rfind('\n', 0, open_start) + 1 #excludes newline
    tag_indent = new_text[open_indent_start : open_start]

    #General Indenting
    spaces_per_indent = int(len(tag_indent) / 2) #the style tag should always be in the 2nd indent tier
    indent_block = ' ' * spaces_per_indent
    
    # Newline + Indenting after every '{' & ';'
    css = css.replace('{', '{\n' + indent_block * 4)
    css = css.replace(';', ';\n' + indent_block * 4)

    # Newline + Indenting before and after every '}'
    css = css.replace('}', '\n' + indent_block * 3 + '}\n\n' + indent_block * 3)

    # Replacing in newly formatted css text
    new_text = new_text[:open_end + 1] + css + new_text[close_start:]

    return new_text

newer_text = css_organize(new_text)

#print(newer_text)
```

```python
# style tags from google classes

def selector_finder(location: list, new_text=new_text):
    '''
    Takes a list of indices in the text where a particular css statement occurs, and
    returns a list of the selectors attributed to it, eg. ".c16" or "h3".

    * Separated from class_finder to avoid redundant code when re-used for lists.
    '''

    selector_list = []
    for index in location:
        close_bracket = new_text.rfind('}', 0, index)
        open_bracket = new_text.find('{', close_bracket, index)
        selector_name = new_text[close_bracket+1:open_bracket]
        selector_name = selector_name.strip()
        selector_list.append(selector_name)

    return selector_list

def class_finder(location: list, new_text=new_text):
    '''
    Takes a list of indices in the text where a particular css statement occurs, and
    returns a list of the class names attributed to it, eg. "c16".
    '''

    selector_list = selector_finder(location, new_text)
    class_list = []
    for selector in selector_list:
        if selector[0] == '.': #finds selectors that are classes specifically, eg ".c16"
            class_list.append(selector[1:]) #appends only the class name, eg "c16"

    return class_list


def style_tagger(text=newer_text):

    #copy text
    new_text = text
    
    ## finding the classes for bold, italics, and underline text

    #the particular css identifiers for text styling
    bold_id = 'font-weight:700'
    italics_id = 'font-style:italic'
    underline_id = 'text-decoration:underline'

    #list of indices where the identifiers occur WITHIN THE CSS
    bold_locations = character_locator(bold_id, new_text)
    italics_locations = character_locator(italics_id, new_text)
    underline_locations = character_locator(underline_id, new_text)

    #list of the identified classes
    bold_classes = class_finder(bold_locations, new_text)
    italics_classes = class_finder(italics_locations, new_text)
    underline_classes = class_finder(underline_locations, new_text)

    """ #For Diagnosis Purposes
    print(f'The bold classes are: {bold_classes}')
    print(f'The italic classes are: {italics_classes}')
    print(f'The underline classes are: {underline_classes}')

    # Because it is possible for some elements to feature multiple classes in their attributes,
    # 1. the solution for now will be to first search every instance of class attributes
    # 2. then search within each class string whether an instance of the stylization tags occur
    # 3. if the stylization class is found, then the corresponding HTML tag should be included outside the current element
    # 4. it may also be necessary to find the type of tag that has the class attribute, in order to identify the closing tag as well
    #   - this above may not be necessary to do this considering these are probably just span tags. """

    #find body tag
    body_index = new_text.find('<body')
    print(f'The body tag starts at index {body_index}.') #Diagnosis

    # All class instances (obsolete):
    class_id = 'class="'
    class_locations = character_locator(class_id, new_text, index_start=body_index)
    
    print(f'The number of instances of tags with class attributes is {len(class_locations)}.') #Diagnosis

    
    #Initialize Conditional Parameters:
    class_att_index = new_text.find(class_id, body_index)
    loop = True

    #Initialize Editing Loop:
    while loop:

        # Loop Breaker
        if class_att_index == -1:
            loop = False
            break

        class_close = new_text.find('"', class_att_index+len(class_id)) #starts search after first quotation mark
        content_string = new_text[class_att_index+len(class_id):class_close] #string excludes final quotation mark
        content_list = content_string.split()

        #reset all flags to false
        bold_flag = False
        italics_flag = False
        underline_flag = False

        #check to see if any classes listed for class instance are for italics, bold, or underline
        for name in content_list:
            if name in italics_classes and not italics_flag:
                italics_flag = True

            if name in bold_classes and not bold_flag:
                bold_flag = True
                
            if name in underline_classes and not underline_flag:
                underline_flag = True

        #if no flag was turned on, move to the next class instance.
        if not (italics_flag or bold_flag or underline_flag):
            class_att_index = new_text.find(class_id, class_att_index + len(class_id))
            continue

        #if a flag or flag combination was turned on, create a string for it

        #create empty strings for tags
        style_open = ""
        style_close = ""
            
        #apply tags to open and closing string
        if italics_flag:
            style_open = "<i>" + style_open
            style_close = style_close + "</i>"
        if bold_flag:
            style_open = "<b>" + style_open
            style_close = style_close + "</b>"
        if underline_flag:
            style_open = "<u>" + style_open
            style_close = style_close + "</u>"

        #finding the "<" character of the opening tag
        open_start = new_text.rfind("<", body_index, class_att_index)

        #finding the ">" character of the opening tag
        open_end = new_text.find(">", class_att_index)

        #finding the type of tag, eg 'span'
        space_index = new_text.find(" ", open_start)
        tag_type = new_text[open_start + 1 : space_index] #excludes both '<' and ' '
        
        #finding the closing tag
        close_index = new_text.find("</" + tag_type + ">", open_end) #closing tag included to avoid possible tag confusion

        #update tags in text to exclude span tags
        if tag_type == 'span':
            new_text = (
                new_text[0:open_start] # excludes span tag
                + style_open 
                + new_text[open_end+1: close_index] #only text between open and close tags
                + style_close 
                + new_text[new_text.find(">",close_index)+1 :] #excludes </span> tag
            )

        else: #in case some of these are <p> tags, which they shouldn't be
            new_text = (
                new_text[0:open_end+1] 
                + style_open 
                + new_text[open_end+1: close_index] #only text between open and close tags
                + style_close 
                + new_text[close_index:]
            )

        #update current index
        class_att_index = new_text.find(class_id, class_att_index + len(class_id))
    
    print(new_text[body_index:])

    return new_text

style_tagger()
```
